```
GA_condition = public_sch_nat['STABR'] == 'GA' #set State Abreviation to Georgia in national Public School Characteristics 2017-2018 CSV file

public_sch_GA = public_sch_nat[GA_condition] #create DataFrame with only Georgia public schools from national Public School Characteristics 2017-2018 CSV file

public_sch_GA.head()
```

```
math_national = pd.read_csv('math-achievement-sch-sy2017-18.csv') #create a DataFrame with only Georgia public schools from national Math achievements for 2017-2018 CSV file

math_national.head()
```

```
GA_condition_math = math_national['STNAM'] == 'GEORGIA' #set state name to Georgia from national Math achievements for 2017-2018 CSV file

math_GA = math_national[GA_condition_math] #create DataFrame with only  national Math achievements for 2017-2018 CSV file

math_GA.head()
```

```
rla_national = pd.read_csv('rla-achievement-sch-sy2017-18.csv') #create DataFrame for national Reading and Language achievements for 2017-2018 CSV file

rla_national.head()
```

```
GA_condition_rla = rla_national['STNAM'] == 'GEORGIA' #set State Name to Georgia for national Reading and Language achievements for 2017-2018 CSV file

rla_GA = rla_national[GA_condition_rla] #create DataFrame with only Georgia public schools from national Reading and Language achievements for 2017-2018 CSV file

rla_GA.head()
```

```
rla_GA.shape
```

```
math_rla_GA = pd.merge(math_GA, rla_GA, how = 'outer', on = 'NCESSCH') #combine DateFrames for Georgia public schools Math and Reading and Language achievements for 2017-2018

math_rla_GA.head()
```

```
math_rla_GA.shape
```

```
public_sch_math_rla_GA = pd.merge(public_sch_GA, math_rla_GA, how = 'outer', on = 'NCESSCH') #combine Georgia Public School Characteristics, Math and Reading and Language achievements for 2017-2018

public_sch_math_rla_GA.head()
```
